## Categorical Analysis

```python
def categoric_data(dataframe, cat_th = 12, car_th=20):
    cat_cols = [col for col in dataframe.columns if str(dataframe.dtypes[col]) in ["category", "bool", "object"]]
    num_but_cat = [col for col in dataframe.columns if str(dataframe.dtypes[col]) in ["int64", "float64"] and dataframe[col].nunique() < cat_th]
    cat_but_car = [col for col in dataframe.columns if str(dataframe.dtypes[col]) in ["category", "object"] and dataframe[col].nunique() > car_th]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    num_cols = [col for col in dataframe.columns if str(dataframe[col].dtypes) in ["int64", "float64"]]
    num_cols = [col for col in num_cols if col not in num_but_cat]
    
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variebles: {dataframe.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}")
    print(f"num_cols: {len(num_cols)}")
    print(f"cat_but_car: {len(cat_but_car)}")
    print(f"num_but_cat: {len(num_but_cat)}")
    
    return cat_cols, num_cols, cat_but_car, num_but_cat
```

```python
cat_cols, num_cols, cat_but_car, num_but_cat = categoric_data(data)
```

```python
# Univariate Analysis
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))
fig.suptitle("Univariate Analysis: Distribution of Key Numerical Variables", fontsize=14)

numeric_features = ['streams', 'in_spotify_playlists', 'in_apple_playlists', 'in_deezer_playlists', 
                    'bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%']

for i, col in enumerate(numeric_features):
    ax = axes[i // 3, i % 3]
    sns.histplot(data[col], bins=30, kde=True, ax=ax, color="blue")
    ax.set_title(f"Distribution of {col}")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Detecting Outliners

```python
from scipy.stats import zscore

numeric_features = data.select_dtypes(include=['int64', 'float64']).columns

z_scores = np.abs(data[numeric_features].apply(zscore))

outlier_mask = (z_scores > 3)
outliers = outlier_mask.sum()
print("Outliers detected in each column:")
print(outliers[outliers > 0])
data_cleaned = data[~outlier_mask.any(axis=1)]
data_cleaned.to_csv("cleaned_dataset.csv", index=False)
```

```python

numeric_features = data.select_dtypes(include=['int64', 'float64']).columns

z_scores = np.abs(data[numeric_features].apply(zscore))
outlier_mask = (z_scores > 3)

data_cleaned = data[~outlier_mask.any(axis=1)]
data_cleaned.to_csv("cleaned_dataset.csv", index=False)

print(f"Original dataset shape: {data.shape}")
print(f"Cleaned dataset shape: {data_cleaned.shape}")
```

### Conclusion

This analysis highlights how different audio features interact and contribute to a song’s overall appeal. The correlation between danceability, energy, and valence suggests that high-energy songs with positive emotions tend to be more engaging. The genre and mode distributions indicate that specific musical styles dominate certain listener preferences. These insights can assist in refining music recommendation algorithms, curating better playlists, and guiding music producers in crafting songs aligned with popular trends. Furthermore, understanding these trends can help streaming services optimize user engagement, ensuring that their curated content resonates with the audience.
